```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
# %matplotlib inline

df = pd.read_csv("./dataset/crimes_against_women_2001-2014.csv")
# print(df.head())
print(df.shape)
# print(df.describe())
# print(df.info())

print(df.isnull().sum())
df = df.dropna()

print(df['STATE/UT'].value_counts())
print(df['Assault on women with intent to outrage her modesty'].value_counts())

```

```python
df[['DISTRICT','Assault on women with intent to outrage her modesty']]
```

```python


# Calculate the total number of cases for each location and year
df_grouped = df.groupby(['STATE/UT', 'DISTRICT', 'Year']).sum()

# Calculate the total number of cases for each location
location_totals = df.groupby(['STATE/UT', 'DISTRICT']).sum()
location_totals['Total Cases'] = location_totals.sum(axis=1)

# Calculate the total number of cases for each year
year_totals = df.groupby(['Year']).sum()
year_totals['Total Cases'] = year_totals.sum(axis=1)

# Calculate the safety score for each location and year
safety_scores = (df_grouped['Rape'] + df_grouped['Kidnapping and Abduction'] + 
                df_grouped['Dowry Deaths'] + df_grouped['Assault on women with intent to outrage her modesty'] +
                df_grouped['Insult to modesty of Women'] + df_grouped['Cruelty by Husband or his Relatives'] + 
                df_grouped['Importation of Girls']) / location_totals['Total Cases']

# Add the safety scores to the df_grouped DataFrame
df_grouped['safety_score'] = safety_scores.values

# Calculate the safety score for each year
year_safety_scores = (year_totals['Rape'] + year_totals['Kidnapping and Abduction'] + 
                      year_totals['Dowry Deaths'] + year_totals['Assault on women with intent to outrage her modesty'] +
                      year_totals['Insult to modesty of Women'] + year_totals['Cruelty by Husband or his Relatives'] + 
                      year_totals['Importation of Girls']) / year_totals['Total Cases']

# Add the safety scores to the year_totals DataFrame
year_totals['safety_score'] = year_safety_scores.values

# Reset the index to make sure it's compatible with the original DataFrame
df_grouped = df_grouped.reset_index()
location_totals = location_totals.reset_index()
year_totals = year_totals.reset_index()

# Print out the safety scores
print(df_grouped[['STATE/UT', 'DISTRICT', 'Year', 'safety_score']])
# print(location_totals[['STATE/UT', 'DISTRICT', 'Total Cases', 'safety_score']])
print(year_totals[['Year', 'Total Cases', 'safety_score']])
```
